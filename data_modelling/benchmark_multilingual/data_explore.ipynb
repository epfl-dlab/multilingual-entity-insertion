# Compare the number of candidates per language

```python
import matplotlib.pyplot as plt
```

```python
df_en['candidates'] = df_en['negative_contexts'].apply(lambda x: x.count("'context'") + 1)
```

```python
df_simple['candidates'] = df_simple['negative_contexts'].apply(lambda x: x.count("'context'") + 1)
```

```python
df_en.hist(column='candidates', bins=100)
# set x and y log scale
plt.xscale('log')
plt.yscale('log')
```

```python
df_simple.hist(column='candidates', bins=100)
# set x and y log scale
plt.xscale('log')
plt.yscale('log')
```

```python
simple_mrr = df_simple['random_rank'].apply(lambda x: 1/x).mean()
en_mrr = df_en['random_rank'].apply(lambda x: 1/x).mean()

print(f'Simple MRR: {simple_mrr}')
print(f'English MRR: {en_mrr}')
```

```python
from ast import literal_eval
for i in range(10):
    sample = df_en[df_en['candidates'] < 10].sample(1)
    print(sample['candidates'].iloc[0])
    print(sample['source_title'].iloc[0])
    print(sample['target_title'].iloc[0])
    candidates = literal_eval(sample['negative_contexts'].iloc[0])
    for c in candidates:
        print(c['context'])
        print('-')
    print('##########')
```
